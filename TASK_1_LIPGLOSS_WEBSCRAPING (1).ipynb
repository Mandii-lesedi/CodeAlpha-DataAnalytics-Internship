```python

#TASK 1 – Web Scraping: Lip‑Gloss Products

#• Scrapes multiple catalogue pages
#• Cleans and structures the data
#• Saves results to CSV
#• Prints quick‑fire summary stats
#• Generates a simple price histogram



!pip install selenium beautifulsoup4 pandas
!apt-get update
!apt install -y chromium-chromedriver

from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time

# Set up Chrome options
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Initialize driver (FIXED: pass options as a keyword arg)
driver = webdriver.Chrome(options=options)  # Removed 'chromedriver' positional arg

url = "https://www.amazon.com/Maybelline-Hyaluronic-Hydrating-Hydrated-Fuller-Looking/dp/B091HZ6TZ8/"
driver.get(url)
time.sleep(2)

# Get product title and price
soup = BeautifulSoup(driver.page_source, 'html.parser')
title = soup.select_one('#productTitle').get_text(strip=True)
price = soup.select_one('.a-price .a-offscreen').get_text(strip=True) if soup.select_one('.a-price .a-offscreen') else 'N/A'

# Click "See all reviews" (if button exists)
try:
    driver.find_element("css selector", 'a[data-hook="see-all-reviews-link-foot"]').click()
    time.sleep(2)
except:
    pass

# Scrape first 5 reviews (for speed)
reviews = [review.text.strip() for review in driver.find_elements("css selector", '.review-text-content')][:5]
driver.quit()

# Save to DataFrame
data = {
    "Title": [title],
    "Price": [price],
    "Top Reviews": [" | ".join(reviews)]  # Combine reviews into one cell
}
pd.DataFrame(data).to_csv("lipgloss_quick.csv", index=False)
print("Done! Output saved to 'lipgloss_quick.csv'")
```
