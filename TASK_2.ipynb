```python
df = pd.read_csv('styles.csv', engine='python', on_bad_lines='skip')
display(df.head())
```

```python
# Check dataset shape (rows, columns)
print("Shape:", df.shape)

# Check data types & missing values
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())

# Summary statistics
df.describe()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='masterCategory', order=df['masterCategory'].value_counts().index)
plt.title("Product Distribution by Master Category")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()

#MEN vs Kids vs Women
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='gender')
plt.title("Product Distribution by Gender")
plt.show()

#top 10 colors that are in fashion profucts
plt.figure(figsize=(12, 6))
top_colors = df['baseColour'].value_counts().head(10)
sns.barplot(x=top_colors.values, y=top_colors.index)
plt.title("Top 10 Colors in Fashion Products")
plt.xlabel("Count")
plt.ylabel("Color")
plt.show()

```
