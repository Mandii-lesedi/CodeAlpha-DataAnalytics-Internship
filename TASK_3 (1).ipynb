```python
df = pd.read_csv('Iris.csv', engine='python', on_bad_lines='skip')

# Load the Iris dataset from Seaborn
iris = sns.load_dataset('iris')
print(iris.head())  # Preview the first 5 rows
print(iris.describe())  # Summary statistics
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(iris['petal_length'], bins=20, kde=True, color='blue')
plt.title('Distribution of Petal Length in Iris Flowers')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()   #histogram of the petal lengths
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris, hue='species', palette='Set1')
plt.title('Sepal Length vs. Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show() # speal length vs Width
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_width', data=iris, palette='Pastel1')
plt.title('Petal Width Distribution by Iris Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
sns.pairplot(iris, hue='species', palette='husl')
plt.suptitle('Pairwise Relationships in Iris Dataset', y=1.02)
plt.show()
```
